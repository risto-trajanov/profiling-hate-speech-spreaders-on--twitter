```python
def visualize_word_count(show_words, df):
    show_words = show_words
    plt.bar(df['words'], df['count'], color='crimson')
    plt.title("Word Count")
    plt.ylabel("Count")
    plt.yscale('log') # optionally set a log scale for the y-axis
    plt.xticks(rotation=90)
    for i, (tag, count) in enumerate(zip(df['words'][:show_words], df['count'][:show_words])):
        plt.text(i, count, f' {count} ', rotation=90,
                 ha='center', va='top' if i < 10 else 'bottom', color='white' if i < 10 else 'black')
    plt.xlim(-0.6, len(df['words'][:show_words])-0.4) # optionally set tighter x lims
    plt.tight_layout() # change the whitespace such that all labels fit nicely
    plt.rcParams["figure.figsize"] = (20,3)
    plt.show()
```
